Análisis

```python
#
# Agrupar por 'Producto' y 'Mes' y luego calcular la suma y el conteo de 'Ausente'
grouped = merged_df2.groupby(['Producto', 'Mes'])['Ausente']

ausente_counts = grouped.sum()

total_counts = grouped.count()

percentage_ausente = (ausente_counts / total_counts) * 100

percentage_ausente = percentage_ausente.reset_index(name='Porcentaje_Ausente')
```
